<h2>Data analysis 

```python
loan_amounts = data['loan_amount'].values
```

```python
mean_loan =np.mean(loan_amounts)
median_loan =np.median(loan_amounts)
std_loan =np.std(loan_amounts)

print(f"Mean: {mean_loan:.2f}, Median: {median_loan:.2f}, Std Dev: {std_loan:.2f}")
```

```python
avg_loan_by_type = data.groupby('loan_type')['loan_amount'].mean()
print(avg_loan_by_type)
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data['loan_amount'], kde=True, color='blue')
plt.title('Distribution of Loan Amounts')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='loan_status', hue='gender', data=data)
plt.title('Loan Status Breakdown by Gender')
plt.show()
```

```python
data['disbursed_date'] = pd.to_datetime(data['disbursed_date'])
data.set_index('disbursed_date').resample('M').size().plot(figsize=(12, 6))
plt.title('Number of Loans Disbursed Over Time (Monthly)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='annual_income', y='loan_amount', hue='loan_type', data=data, alpha=0.5)
plt.title('Annual Income vs. Loan Amount')
plt.show()
```
